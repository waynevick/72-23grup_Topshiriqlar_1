```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

# Ma'lumotlar (Misol tariqasida)
X = np.array([[70], [80], [85], [90], [95], [100]])  # Uy o'lchami
y = np.array([15000, 20000, 25000, 45000, 50000, 60000])  # Narxlar

# Qaror daraxti modelini yaratish
model = DecisionTreeRegressor(max_depth=3)
model.fit(X, y)

# Daraxtni chizish
plt.figure(figsize=(10, 6))
plot_tree(model, filled=True, feature_names=["House Size"], rounded=True)
plt.show()
```
